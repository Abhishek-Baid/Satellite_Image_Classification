## Loading Images : 
Function used to load images as numpy arrays. Open cv2 is being used to read the images and store in array of images for training and a combination of image name and label(class) of image. 

```python
#ROOT_DIR='D:\\Abhi\\cloudxlab\\DS PGP\\CapstoneProject\\EuroSAT_RGB\\EuroSAT_RGB'
ROOT_DIR='C:\\Users\\Abhi\\Jupyter_notebooks\\Cloudxlab DS PGP\\MachineLearning\\CapstoneProject_Satellite_Image_Classification\\EuroSAT_RGB\\EuroSAT_RGB\\'
os.listdir(ROOT_DIR)

def load_images_from_directory(ROOT_DIR):
    images=[]
    labels=[]
    image_names=[]
    i=0
    for dir_name in os.listdir(ROOT_DIR):
        class_name = dir_name.split("\\")[-1]
        print(class_name)
        sub_dir_name = os.path.join(ROOT_DIR,dir_name)
        print(sub_dir_name)

        if os.path.isdir(sub_dir_name):    
            for img in os.listdir(sub_dir_name):
                if img.endswith(('.jpg', '.jpeg', '.png')):                    
                    image_read = cv2.imread(os.path.join(ROOT_DIR,dir_name,img))
                    #print("{} image shape is : {}".format(img, image_read.shape))
                    img_cvt_color = cv2.cvtColor(image_read, cv2.COLOR_BGR2RGB)
                    
                    image_names.append(img)
                    images.append(img_cvt_color)
                    labels.append(class_name)
#        i+=1
#        if i==1:
#            break
    return np.array(images), np.column_stack(np.array(labels), np.array(image_names))
```

## Metrics Calculation:
As we are dealing with a classification problem, metrics such as precision, recall and f1 score are a good measure other than accuracy. Function to get precision, recall, f1 score as well as the confusion matrix when actual value, and predicted value is passed.

```python
from sklearn.metrics import precision_score, recall_score, f1_score, confusion_matrix
#from sklearn.metrics import plot_confusion_matrix

def precision_recall_f1_value(actual_value, predicted_value, classes, average_type='binary'):
    precision_value = precision_score(actual_value, predicted_value, average=average_type)
    recall_value = recall_score(actual_value, predicted_value, average=average_type)
    f1 = f1_score(actual_value, predicted_value, average=average_type)
    conf_matrix = confusion_matrix(actual_value, predicted_value)
    print("Precision Value : " + str(precision_value))
    print("Recall value : " + str(recall_value))
    print("F1 Score : " + str(f1))
    print("Following is the confusion matrx for this model : ")
    print(conf_matrix)
    plt.matshow(conf_matrix, cmap=plt.cm.gray)
    
    plt.imshow(conf_matrix, interpolation='nearest', cmap=plt.cm.gray)
    plt.title('Confusion Matrix')
    plt.colorbar()
    classes=classes
    
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes, rotation=0)
    
    plt.xlabel('Predicted Labels')
    plt.ylabel('True Labels')
    plt.show()
    
    #plot_confusion_matrix(estimator, test_features, test_label, values_format='.5g', display_labels=class_names)
    #return precision_value, recall_value
    
```

```python
def perform_ipca(X_train, X_test, n_components=None, ipca_batch_size=1000 ):

    ipca_batch_size=ipca_batch_size

    ipca = IncrementalPCA(n_components=n_components, batch_size=ipca_batch_size)

    #Incrementally fitting data in PCA batches

    for i in range(0, len(X_train), ipca_batch_size):
        X_batch = X_train[i:i+ipca_batch_size, :]
        ipca.partial_fit(X_batch)

        #Transform the training Data
    X_train_reduced = ipca.transform(X_train)
    X_test_reduced = ipca.transform(X_test)
    
    return X_train_reduced, X_test_reduced
```

```python
#Function to run any model , pass the model params and other params from outside


def perform_randomized_search(model, param_grid, X_train, y_train, n_iter=5, cv=5, scaling=True, random_state=83):
    """
    Perform randomized search for hyperparameter tuning
    
    Parameters:
    - model : ML algorithm
    - param_grid : Parameter Grid for parameter distribution from which random combinations will be taken.
    - X_train : Feature data
    - y_train : Label
    - n_iter: Number of parameter settings that are sampled.
    - cv : Number of cross validation folds
    - randome_state : defaulted to 83
    """
    
    random_search = RandomizedSearchCV(model, param_distributions=param_grid, n_iter=n_iter, cv=cv, random_state=random_state)
    
    scaler = StandardScaler()
    if scaling:
        X_train_model = scaler.fit_transform()
    else:
        X_train_model = X_train
    
    #Fit the random_search to data
    random_search.fit(X_train_model, y_train)
    
    return random_search.best_estimator_



def perform_grid_search(model, param_grid, X_train, y_train, scoring=None, cv=5, scaling=True, random_state=83):
    """
    Perform randomized search for hyperparameter tuning
    
    Parameters:
    - model : ML algorithm
    - param_grid : Parameter distribution.
    - X_train : Feature data
    - y_train : Label
    - scoring 
    - cv : Number of cross validation folds
    - scaling : If scaling is required.
    - random_state : defaulted to 83
    """
    
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid, scoring=scoring, cv=cv, verbose=True, return_train_score=True)
    
    scaler = StandardScaler()
    if scaling:
        X_train_model = scaler.fit_transform(X_train)

    else:
        X_train_model = X_train
    
    #Fit the random_search to data
    grid_search.fit(X_train_model, y_train)
    
    return grid_search.best_estimator_
```
